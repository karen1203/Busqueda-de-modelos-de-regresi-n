```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('cuentas.xlsx')
```

```python
df=df[['precio','enganche','descuento','semana','monto_financiado','costo_total','monto_accesorios','status','fraude','inversion','pagos_realizados','reautorizacion','porc_enganche','score_buro']]
```

```python
df.isnull().sum()
```

```python
df['porc_enganche']=df['enganche'].fillna(0)
```

Precio vs Enganche

```python
Vars_Indep= df[['enganche']]
Var_Dep= df['precio']
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
type(model)
```

```python
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
model.__dict__
```

```python
model.score(Vars_Indep,Var_Dep)
```

```python
y_pred= model.predict(X=df[['enganche']])
y_pred
```

```python
sns.scatterplot(x='enganche', y='precio', color="blue", data=df)
```
